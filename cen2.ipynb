```python
base_dir = "/home/tong/Desktop/MoviesSimon/"
fname = "/home/tong/Desktop/MoviesSimon/movie-1442_4/movie-1442_4_MMStack_Pos0.ome.tif"
tracker = Cen2Tracker(fname, base_dir=base_dir, verbose=True, force_metadata=False,
                      json_discovery=True, clean_store=True)
tracker.metadata
```

```python
traj_name = "peaks_real"
tkw = TrackersWidget(tracker, traj_name,
                     xaxis='t', yaxis='x_proj',
                     scale_x=1, scale_y=1)
tkw.show()
```
